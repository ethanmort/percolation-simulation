# Percolation Model Results
Plots results (`percolation_results.csv`) generated from classical percolation model

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("percolation_results.csv")
data.head()
```

## Plotting

### Percolation Curve

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(data['p'], data['chance'])
axes.set_xlabel(r"$p$")
axes.set_ylabel("Probability of Percolation")
fig.tight_layout()
```

### Max Cluster Size

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(data['p'], data['max_cluster'])
axes.set_xlabel(r"$p$")
axes.set_ylabel("Max Cluster Size")
fig.tight_layout()
```
