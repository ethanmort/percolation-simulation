# Percolation Model Results
Plots results (`percolation_results.csv`) generated from classical percolation model

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../output/percolation_results.csv")
data.head()
```

## Plotting

### Percolation Curve

```python
fig, axes = plt.subplots(nrows=1, ncols=1)

for (geometry, h, w), subset in data.groupby(["geometry", "h", "w"]):
    axes.plot(subset['p'], subset['chance'], label=f"{geometry}, h={h}, w={w}")

axes.set_xlabel(r"$p$")
axes.set_ylabel("Probability of Percolation")
axes.legend()
fig.tight_layout()
```

### Max Cluster Size

```python
fig, axes = plt.subplots(nrows=1, ncols=1)

for (geometry, h, w), subset in data.groupby(["geometry", "h", "w"]):
    axes.plot(subset['p'], subset['max_cluster'], label=f"{geometry}, h={h}, w={w}")


axes.set_xlabel(r"$p$")
axes.set_ylabel("Max Cluster Size")
axes.legend()
fig.tight_layout()
```
